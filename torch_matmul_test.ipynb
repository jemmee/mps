```python
import torch
import time

# 1. Set the device to 'mps' (Metal Performance Shaders)
device = torch.device("mps")

# 2. Create two large matrices on the GPU
size = 8192
a = torch.randn(size, size, device=device)
b = torch.randn(size, size, device=device)

print(f"Running MatMul on M1 GPU ({size}x{size})...")

# 3. Benchmark
start_time = time.time()
# The @ symbol performs matrix multiplication
result = a @ b

# Syncing is important to get an accurate time on Apple Silicon
torch.mps.synchronize() 
print(f"Time on M1 GPU: {time.time() - start_time:.4f} seconds")
```
