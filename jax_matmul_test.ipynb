```python
import jax
import jax.numpy as jnp
import time

# JAX automatically detects the Metal GPU if jax-metal is installed
print(f"Backend: {jax.default_backend()}")
print(f"Device: {jax.devices()[0]}")

size = 8192
key = jax.random.PRNGKey(0)
a = jax.random.normal(key, (size, size))
b = jax.random.normal(key, (size, size))

start = time.time()
# Use block_until_ready() to wait for the GPU to finish
result = jnp.dot(a, b).block_until_ready()
print(f"JAX on M1 GPU: {time.time() - start:.4f} seconds")
```
